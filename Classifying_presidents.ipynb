```python
#Function to extract the answers from chatGPT
def generate_text(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        #model="gpt-3.5-turbo",
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.2
        )
    return response.choices[0].text.strip()
```

```python
# Function to extract the value a csv file
def copy_first_column(filename):
    with open(filename, 'r', encoding='mbcs') as csv_file:
        reader = csv.reader(csv_file)
        copied_values = [row[0] for row in reader]
        return copied_values
```

```python
# Creating the prompt of the country leaders
prompts_vanilla = []
prompts_FK = []
prompts_RILE = []
prompts_speech = []
prompts_policy = []
prompts_self = []

for leader in country_leaders[1:]:
    prompt_vanilla = "In one word, based on the vanilla method, was "+leader+" influenced by left, center-left, center, center-right, or right-leaning ideologies?"
    prompt_FK = "In one word, based on the Franzmann and Kaiser method, was "+leader+" influenced by left, center-left, center, center-right, or right-leaning ideologies?"
    prompt_RILE = "In one word, based on the RILE of the party manifesto group method, was "+leader+" influenced by left, center-left, center, center-right, or right-leaning ideologies?"
    prompt_speech = "In one word, based on the content speech, was "+leader+" influenced by left, center-left, center, center-right, or right-leaning ideologies?"
    prompt_policy = "In one word, based on policy implemented, was "+leader+" influenced by left, center-left, center, center-right, or right-leaning ideologies?"
    prompt_self = "In one word, based on their self-identifcation, was "+leader+" influenced by left, center-left, center, center-right, or right-leaning ideologies?"
    
    prompts_vanilla.append(prompt_vanilla)
    prompts_FK.append(prompt_FK)
    prompts_RILE.append(prompt_RILE)
    prompts_speech.append(prompt_speech)
    prompts_policy.append(prompt_policy)
    prompts_self.append(prompt_self)
```
